### Entrainement

```python
N = 1000    # Hyper parameters (N -> num epoch & lr -> learning rate)
lr = 0.1
```

```python
w = random.random()   # Re initialisation des parametres pour entrainer
b = random.random()
dw_history = []   # Sauvegarde les données d'apprentissage des différent parametres
db_history = []
for epoch in range(N):
    dw = 0    # initalisation des gradiant a 0 pour chaque epoch
    db = 0

    for i in range(m):
        y_pred = w * x[i] + b   # Predire y pour chaque donnée

        dw = dw + (y_pred - y[i]) * x [i]  # Calcule de l'écartement et donc des gradients
        db = db + (y_pred - y[i])
    
    dw = 2 * dw/m   # Moyenne des gradients
    db = 2 * db/m
    dw_history.append(dw)
    db_history.append(db)

    w = w - lr*dw   # Modification des parametres en fonction des gradients
    b = b - lr*db
```

### Validation et verification

```python
plt.plot(dw_history) # Affiche les courbes d'apprentissages
plt.plot(db_history)
(w, b)
```
